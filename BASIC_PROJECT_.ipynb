```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import cv2
import tensorflow as tf
from sklearn.model_selection import train_test_split
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras import layers, models
from tensorflow.keras.layers import MaxPooling2D,Flatten,Dense
from tensorflow.keras.layers import Conv2D
```

```python
# Main dataset folder
dataset_path = r"C:\Users\SURFACE\Desktop\DATA SCIENCE WITH AI AND ML\MY PROJECTS\PROJECTS\cnn-building\DATASET_FILE_FOR CNN_BULIDING"

# Training folder
train_dir = os.path.join(dataset_path, "seg_train") # IN THIS TRAIN_DIR IT USED FOR TRAINING FOLDER 

# Testing folder
test_dir = os.path.join(dataset_path, "seg_test")#IN THIS TEST_DIR IT USED FOR TESTING FOLDER 

# Prediction folder
pred_dir = os.path.join(dataset_path, "seg_pred")#IN THIS PRED_DIR IT USED FOR PREDICTION OF ANY DATA SET.
```

```python
train_data = tf.keras.utils.image_dataset_from_directory(
    train_dir,
    validation_split=0.2,
    subset="training",
    seed=42,
    image_size=(224,224),
    batch_size=32
)
```

```python
validation_data = tf.keras.utils.image_dataset_from_directory(
    train_dir,
    validation_split=0.2,
    subset="validation",
    seed=42,
    image_size=(224,224),
    batch_size=32
)
```

```python
test_data = tf.keras.utils.image_dataset_from_directory(
    test_dir,
    image_size=(224,224),
    batch_size=32
)
```

```python
normalization_layer = layers.Rescaling(1./255)

train_data = train_data.map(lambda x, y: (normalization_layer(x), y))
validation_data = validation_data.map(lambda x, y: (normalization_layer(x), y))
test_data = test_data.map(lambda x, y: (normalization_layer(x), y))
```

```python
model = models.Sequential([

    layers.Input(shape=(224,224,3)),

    layers.Conv2D(32,(3,3),activation='relu'),
    layers.MaxPooling2D(),

    layers.Conv2D(64,(3,3),activation='relu'),
    layers.MaxPooling2D(),

    layers.Conv2D(128,(3,3),activation='relu'),
    layers.MaxPooling2D(),

    layers.Flatten(),

    layers.Dense(128,activation='relu'),

    layers.Dense(6,activation='softmax')

])
```

```python
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
history = model.fit(
    train_data,
    validation_data=validation_data,
    epochs=15
)
```

```python
loss, accuracy = model.evaluate(test_data)

print("Test Accuracy :", accuracy)
```
